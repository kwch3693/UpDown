```python
def pick_strongest(features, corr_matrix, correlation_level):
    strongest = {}
    while features.size > 0:
        f = features.abs().idxmax()
        corrlations = corr_matrix[f]
        correlated_features = corrlations[corrlations > correlation_level].index
        inter = pd.Series(list(set(strongest).intersection(set(correlated_features))))

        if inter.size == 0:
            strongest[f] = features[f]
        features = features.drop(f)
    return strongest  
```

```python
corr_features = corr[np.abs(corr)>0.1].sort_values()
d = pick_strongest(corr_features, corr_matrix, 0.75)
d = pd.Series(d)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
def plot_roc(classifier, y_true, y_score, title):
    
    # Compute ROC curve and ROC area for each class
    truePostiveRate, falsePostiveRate, threshold = roc_curve(y_true, y_score)
    roc_auc = roc_auc_score(y_true, y_score)
    plt.figure()
    lw = 2
    plt.plot(truePostiveRate, falsePostiveRate, color='darkorange',
             lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.title(title)

    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc="lower right")
    plt.show()
```

```python
y_score = rand_fs.predict_proba(X_test)[:, 1]
plot_roc(rand_fs, y_test, y_score, "Random forest with Feature Selection ROC")
```

```python
y_score = rand_no_fs.predict_proba(X_test)[:, 1]
plot_roc(rand_no_fs, y_test, y_score, "Random forest without Feature Selection ROC")
```
